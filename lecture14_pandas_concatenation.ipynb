```python
data1 = {'Name': ['Zaid', 'Junaid', 'Saqeeb', 'Azan'],
         'Age': [27, 24, 22, 32],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']}

data2 = {'Name': ['Abdullah', 'Aqsan', 'Dayan', 'Sayam'],
         'Age': [17, 14, 12, 52],
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']}

df1 = pd.DataFrame(data1, index=[0, 1, 2, 3])

df2 = pd.DataFrame(data2, index=[0, 1, 2, 3])

display(df1, df2)
```

```python
# 2️⃣ Merging (merge()):
# this is used when there is a common column or key baased on whch we merge the dataframes



data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Ali', 'Zubair', 'Sahil', 'Suhaib'], 
        'Age':[27, 24, 22, 32],} 

data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 

data3 = {#'key': ['K0', 'K1', 'K2', 'K3'],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
         'Mobile Number':['00000', '11111', '66666', '999999'], 
        'Status':['Active', 'Offline', 'Dropped', 'Geraduated']} 

df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2) 
df2=  pd.DataFrame(data3)
 
print(df, "\n\n", df1,"\n\n", df2) 

res = pd.merge(df, df1)
 
res2=pd.merge(res,df2)
display(res2)
```

```python
import pandas as pd
import numpy as np

data = {'Name': ['Ali', 'Sara', 'Zain', np.nan, 'Ayesha'],
        'Age': [25, np.nan, 30, 22, np.nan],
        'Salary': [50000, 60000, np.nan, 80000, 70000]}

df = pd.DataFrame(data)
```

```python
display(df)

print(df.isna())  

# Missing values count
print(df.isna().sum())  
```

```python
df_cleaned=df.dropna()
print(df_cleaned)
```

```python
df_cleaned=df.dropna(subset=['Salary'])
print(df_cleaned)
```

```python
df_filled=df.fillna(0)
print(df_filled)
```
